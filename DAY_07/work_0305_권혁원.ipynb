```python
filteredDF = filteredDF.drop(columns=['CRIM', 'ZN', 'CHAS', 'RAD', 'DIS', 'PTRATIO', 'B']).copy()
```

```python
filteredDF.isna().sum()
```

```python
filteredDF = filteredDF.dropna()
```

```python
filteredDF.isna().sum()
```

```python
for col in filteredDF.columns:
    print(filteredDF[col].unique())
```

```python
feature = filteredDF[filteredDF.columns[:-1]]
target = filteredDF[filteredDF.columns[-1]]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(feature, target,
                                                    test_size=0.2,
                                                    random_state=719)
```

```python
X_train.shape, y_train.shape
```

```python
from sklearn.linear_model import LinearRegression

lin_model = LinearRegression()
lin_model.fit(X_train, y_train)
```

```python
# ë‚˜ì˜ì§€ ì•Šì€ë°..?
# ê³¼ì œ ë!
lin_model.score(X_train, y_train), lin_model.score(X_test, y_test)
```

```python
# ìŠ¤ì¼€ì¼ë§
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)

scaled_X_train = scaler.transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

```python
# ë˜‘ê°™ë‹¤
lin_model = LinearRegression()
lin_model.fit(scaled_X_train, y_train)

lin_model.score(scaled_X_train, y_train), lin_model.score(scaled_X_test, y_test)
```

```python
# íŠœë‹í•´ì•¼ê² ì§€?
from sklearn.linear_model import Ridge, Lasso, ElasticNet
from sklearn.model_selection import GridSearchCV
```

```python
# ì•ˆí•´ ì‹¯íŒ”
models = [Ridge(), Lasso(), ElasticNet()]
param = {'alpha':[0.001, 0.01, 0.1, 1, 10, 20, 30],
         'max_iter':[1000],
         'random_state':[719]}

def imp_gscv(model):
    gscv = GridSearchCV(model, param_grid=param, 
                        return_train_score=True,
                        cv=5)
    gscv.fit(scaled_X_train, y_train)
    result = gscv.cv_results_

    return result
```

```python
result_list = []
for model in models:
    result_list.append(imp_gscv(model))
```
