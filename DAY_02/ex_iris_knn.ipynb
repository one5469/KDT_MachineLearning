#### 목표: 붓꽃 품종 식별 모델
- 데이터 : 3개 품종을 가진 iris.csv
- 피쳐 : sepal_length, sepal_width
- 타겟 : species
- 유형 : 지도학습 / 분류

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
import pandas as pd

irisDF = pd.read_csv('../data/iris.csv')
irisDF
```

```python
irisDF.plot(kind='scatter', x='sepal_width', y='sepal_length')
```

```python
plt.figure()
for specie in irisDF['species'].unique():
    plt.scatter(irisDF[irisDF['species'] == specie]['sepal_width'], irisDF[irisDF['species'] == specie]['sepal_length'])
```
